```python
#sending respond to establish connection
respond=requests.get(f'https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={date}#')
```

```python
soup = BeautifulSoup(respond.text,'lxml')
```

```python
championships = soup.find_all('div',{'class':'matchCard'})
```

```python
def get_matches(matches_details,championships,ind):
  all_matches = championships[ind].find_all('div',{'class':'item finish liItem'})
  CS_title = championships[ind].find('h2').text.strip()
  num_of_matches = len(all_matches)
  for i in range(num_of_matches):
      #get teams name
      team_a = all_matches[i].find('div',{'class':'teamA'}).text.strip()
      team_b = all_matches[i].find('div',{'class':'teamB'}).text.strip()
      #get score
      match_result = all_matches[i].find('div',{'class':'MResult'}).find_all('span',{'class':'score'})
      score = f'{match_result[0].text.strip()} - {match_result[1].text.strip()}'
      #get match time
      time = all_matches[i].find('div',{'class':'MResult'}).find('span',{'class':'time'}).text.strip()
      matches_details.append({'البطولة':CS_title,'الفريق الاول':team_a,'الفريق الثاني':team_b,'ميعاد المباراه':time,'النتيجة':score})
```

```python
matches_details = []
for i in range(len(championships)):
  get_matches(matches_details,championships,i)
```

```python
matches_details
```

```python
import pandas as pd

df = pd.DataFrame(matches_details)
df
```

```python
df.to_csv('maches_data.csv')
```

```python
def scrap_data(dte,dte1):
  respond=requests.get(f'https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={dte}/{dte1}/2024#')
  soup = BeautifulSoup(respond.text,'lxml')
  championships = soup.find_all('div',{'class':'matchCard'})
  matches_details = []
  for i in range(len(championships)):
    get_matches(matches_details,championships,i)
```

using this url :           https://books.toscrape.com/catalogue/page-4.html

make csv file with book name, price, availability, and image url

```python
url= "https://books.toscrape.com/catalogue/page-4.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
tables = soup.find_all('li',{'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
books = []
for i in range(len(tables)):
    title = tables[i].find('h3').find('a')['title']
    price = tables[i].find('p', class_='price_color').text.strip().replace('Â£', '')
    availability = tables[i].find('p', class_='instock availability').text.strip()
    image_url = 'https://books.toscrape.com/' + tables[i].find('div', class_='image_container').find('img')['src'].replace('../', '')
    books.append([title, price, availability, image_url])


df = pd.DataFrame(books, columns=['Title', 'Price(£)', 'Availability', 'Image URL'])

```
